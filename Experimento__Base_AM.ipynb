```python
import matplotlib.pyplot as plt
import seaborn as sns

# Lista das colunas de promo√ß√µes
promocao_cols = [
    'Q44_1 - Pensando em promo√ß√µes que a Rommanel poderia fazer, quais s√£o as mais atraentes para voc√™ Compre e ganhe outros produtos',
    'Q44_2 - Pensando em promo√ß√µes que a Rommanel poderia fazer, quais s√£o as mais atraentes para voc√™ Pontos em compras/ junte e troque',
    'Q44_3 - Pensando em promo√ß√µes que a Rommanel poderia fazer, quais s√£o as mais atraentes para voc√™ Cashback',
    'Q44_4 - Pensando em promo√ß√µes que a Rommanel poderia fazer, quais s√£o as mais atraentes para voc√™ Sorteio de viagem/ casa/ carro'
]

# Contar as men√ß√µes para cada promo√ß√£o
contagem = {}
for col in promocao_cols:
    count = df[col].dropna().count()
    if 'Compre e ganhe' in col:
        nome = 'Compre e ganhe outros produtos'
    elif 'Pontos em compras' in col:
        nome = 'Pontos em compras / junte e troque'
    elif 'Cashback' in col:
        nome = 'Cashback'
    elif 'Sorteio de viagem' in col:
        nome = 'Sorteio de viagem / casa / carro'
    else:
        nome = col
    contagem[nome] = count

labels = list(contagem.keys())
values = list(contagem.values())

plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
palette = sns.color_palette("coolwarm", len(labels))

barplot = sns.barplot(x=values, y=labels, palette=palette)

# Adicionar valores no topo das barras
for i, v in enumerate(values):
    barplot.text(v + 0.5, i, str(v), color='black', va='center', fontweight='bold')

plt.title('Promo√ß√µes Mais Atraentes para os Respondentes', fontsize=18, weight='bold')
plt.xlabel('N√∫mero de Men√ß√µes', fontsize=14)
plt.ylabel('Tipo de Promo√ß√£o', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 4. Celebridades/Cantores mais citados
celebridades_cols = [
    'Q44_1 - Pensando em promo√ß√µes que a Rommanel poderia fazer, quais s√£o as mais atraentes para voc√™ Compre e ganhe outros produtos',
    'Q44_2 - Pensando em promo√ß√µes que a Rommanel poderia fazer, quais s√£o as mais atraentes para voc√™ Pontos em compras/ junte e troque',
    'Q44_3 - Pensando em promo√ß√µes que a Rommanel poderia fazer, quais s√£o as mais atraentes para voc√™ Cashback',
    'Q44_4 - Pensando em promo√ß√µes que a Rommanel poderia fazer, quais s√£o as mais atraentes para voc√™ Sorteio de viagem/ casa/ carro',
    'Q44_5 - Que influencers ou celebridades voc√™ gosta? Escreva ( para toda a amostra) - Cantores/ m√∫sicos:'
]

celebridades = df[celebridades_cols].values.flatten()
celebridades = [str(x).strip().lower() for x in celebridades if pd.notnull(x)]
contagem = Counter(celebridades)
mais_citados = contagem.most_common(10)

print("\nüé§ Cantores/Celebridades mais citados:")
for nome, qtd in mais_citados:
    print(f"- {nome.title()}: {qtd} men√ß√µes")

# Gr√°fico de barras
nomes = [nome.title() for nome, _ in mais_citados]
valores = [qtd for _, qtd in mais_citados]

plt.figure(figsize=(10,6))
sns.barplot(x=valores, y=nomes, palette='magma')
plt.title('üé§ Cantores e Celebridades Mais Citados')
plt.xlabel('N√∫mero de Men√ß√µes')
plt.ylabel('Nome')
plt.tight_layout()
plt.show()
```
